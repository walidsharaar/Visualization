# Map Chart

is used to show items on a background that is often, but not always, geographical Maps can contain interactive shapes or display markers of different types on an image or map background

```python
# import library

import plotly.express as px
```

```python
df= px.data.carshare()
df.head()
```

# Map based on latitude and longitude


```python


fig = px.scatter_mapbox(df, 
                        lat="centroid_lat", 
                        lon="centroid_lon",     
                        color="peak_hour", 
                        size="car_hours",
                        size_max=15, 
                        zoom=10,
                       mapbox_style="carto-positron")
fig.show()
```
